**Density-Based Spatial Clustering of Applications with Noise**

This notebook demonstrates DBSCAN clustering algorithm, which excels at finding clusters of arbitrary shape and automatically detecting outliers in datasets with varying densities.

### ðŸ“š Step 1: Import Essential Libraries
Loading the required packages for data manipulation, visualization, and density-based clustering

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

### ðŸŒ™ Step 2: Create Synthetic Dataset & Visualize
Generating the "two moons" dataset - perfect for demonstrating DBSCAN's ability to handle non-spherical clusters

```python
from sklearn.datasets import make_moons

X, y = make_moons(n_samples = 250, noise = 0.05)

plt.scatter(X[:, 0], X[:, 1], c = y)
plt.show()
```

#### ðŸ” **Observation: Two Moons Dataset**

**Key Insights from the Initial Data:**
- ðŸŒ™ **Non-Spherical Shape**: Two crescent-shaped clusters that would challenge K-means
- ðŸŽ¯ **Perfect for DBSCAN**: Complex shapes where density-based clustering excels
- ðŸŽ¨ **Color Coding**: Orange and blue represent the two true clusters
- ðŸ“Š **Dataset Characteristics**:
  - **250 samples** with **low noise** (0.05)
  - **Interleaved structure** - clusters curve around each other
  - **Uniform density** within each moon shape
- âš¡ **Algorithm Challenge**: Traditional clustering methods struggle with this shape
- âœ… **DBSCAN Advantage**: Can detect arbitrary cluster shapes based on density

### âš–ï¸ Step 3: Feature Scaling using Standardization
Normalizing features to ensure consistent distance calculations for the density-based algorithm

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

X_scaled
```

### ðŸ¤– Step 4: Train DBSCAN Model
Implementing density-based clustering with epsilon (neighborhood radius) parameter

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps = 0.3)

dbscan.fit(X_scaled)

dbscan.labels_
```

#### ðŸŽ¯ **Parameter Optimization Note**
**Silhouette scoring** can help optimize DBSCAN hyperparameters:
- **eps (epsilon)**: Neighborhood radius for density calculation
- **min_samples**: Minimum points required to form a dense region
- Higher silhouette scores indicate better cluster separation and cohesion

### ðŸ“ˆ Step 5: Final Visualization
Displaying DBSCAN clustering results with noise points identification

```python
plt.scatter(X[:, 0], X[:, 1], c = dbscan.labels_)
plt.show()
```

#### ðŸ” **Observation: DBSCAN Clustering Results**

**Key Insights from the Final Clustering:**
- ðŸŽ¯ **Perfect Shape Recognition**: DBSCAN successfully identified both crescent-shaped clusters
- ðŸŽ¨ **Cluster Labels**: Different colors represent distinct clusters found by the algorithm
- ðŸ” **Noise Detection**: Any outlier points would be labeled as -1 (noise)
- âœ… **Algorithm Success**: 
  - Handled **non-spherical clusters** effectively
  - Maintained **cluster integrity** despite complex shapes
  - **No pre-specification** of cluster count required
- ðŸ“Š **Comparison with True Labels**: Results should closely match the original two-moon structure
- âš¡ **DBSCAN Advantages Demonstrated**:
  - Robust to **arbitrary cluster shapes**
  - Automatic **noise point detection**
  - **Density-based** approach works well for this dataset
