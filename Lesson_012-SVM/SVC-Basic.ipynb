# __`Support Vector Classfication `__

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create synthetic data
from sklearn.datasets import make_classification

X, y = make_classification(n_samples = 1000, n_features = 2, n_classes = 2, n_clusters_per_class = 2, n_redundant = 0, random_state = 42)

X, y
```

```python
# Plot Data

sns.scatterplot(x = pd.DataFrame(X)[0], y = pd.DataFrame(X)[1], hue = y)
plt.show()
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 10)

X_train.shape, X_test.shape
```

```python
# Model Train

from sklearn.svm import SVC
svc = SVC(kernel='linear')
svc.fit(X_train, y_train) 
```

```python
svc.coef_, svc.intercept_    # Intercept only find in linear kernel becuase line is straight else other kernel line is curved
```

```python
# Prediction 
y_pred = svc.predict(X_test)
y_pred
```

```python
# Performance Metrics

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("Accuracy Score is :", accuracy_score(y_test, y_pred), '\n')
print(confusion_matrix(y_test, y_pred), '\n')
print(classification_report(y_test, y_pred), '\n')
```

## __`HyperParameter Tuning With SVC`__

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold

c = [0.1, 1, 10, 100, 1000]
gamma = [1, 0.1, 0.01, 0.001, 0.0001]
kernel = ['rbf', 'poly', 'linear', 'sigmoid']

param = dict(C = c, gamma = gamma, kernel = kernel)

print(param)
```

```python
grid = GridSearchCV(estimator = SVC(), param_grid = param, refit = True, verbose = 3, n_jobs = -1)

grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
# Prediction
y_pred = grid.predict(X_test)
print("Accuracy Score is :", accuracy_score(y_test, y_pred), '\n')
print(confusion_matrix(y_test, y_pred), '\n')
print(classification_report(y_test, y_pred), '\n')
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Train
svc = SVC(kernel="rbf", probability=True)
svc.fit(X_train, y_train)

# Predict probabilities
y_probs = svc.predict_proba(X_test)[:, 1]

# ROC AUC score
auc = roc_auc_score(y_test, y_probs)

# ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_probs)
plt.plot(fpr, tpr, label=f"SVC (AUC = {auc:.2f})")
plt.plot([0,1], [0,1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```
