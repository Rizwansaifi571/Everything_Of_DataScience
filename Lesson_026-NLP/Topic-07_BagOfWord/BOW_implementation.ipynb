# Bag of Words (BOW) - Text Vectorization

## 📌 What is Bag of Words?
**Bag of Words** is a text representation technique that converts text documents into numerical feature vectors by counting word occurrences, ignoring grammar and word order.

### How It Works:
1. **Create Vocabulary**: Collect all unique words from the corpus
2. **Count Occurrences**: For each document, count how many times each word appears
3. **Generate Vector**: Create a numerical vector where each position represents a word's frequency

---

## 🎯 Why BOW Matters
| Advantage | Description |
|-----------|-------------|
| **Simple & Intuitive** | Easy to understand and implement |
| **Effective Baseline** | Works well for many text classification tasks |
| **Language Independent** | Works with any language |
| **Fast Processing** | Computationally efficient |

### Limitations:
- ❌ **Loses Context**: Ignores word order and grammar
- ❌ **Sparse Matrices**: Creates large vectors with many zeros
- ❌ **No Semantics**: "good" and "great" treated as completely different

## 🔧 Today's Task
We'll build a **SMS Spam Classifier** using BOW:
1. Load SMS spam dataset
2. Clean and preprocess text
3. Create BOW representation
4. Explore N-grams for better context

## 📂 Step 1: Load Dataset
Loading the SMS Spam Collection dataset with spam/ham labels.

```python
import pandas as pd 

messages = pd.read_csv(
    "SMSSpamCollection.csv", 
    sep = '\t', 
    names = ['label', 'message']
)

messages
```

## 🧹 Step 2: Import Preprocessing Tools

**What we need:**
- `re` - Remove special characters and numbers
- `stopwords` - Remove common words like "the", "is", "at"
- `word_tokenize` - Split text into words
- `WordNetLemmatizer` - Reduce words to base form ("running" → "run")

Preprocessing reduces noise and improves model accuracy.

```python
# Data Cleaning and preprocessing

import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

from nltk.stem import WordNetLemmatizer
lemma = WordNetLemmatizer()
```

## 🔄 Step 3: Create Cleaned Corpus

**Preprocessing Pipeline:**
1. `re.sub('[^a-zA-Z]', ' ', text)` - Keep only letters
2. `lower()` - Convert to lowercase
3. `word_tokenize()` - Split into words
4. `lemmatize()` - Convert to base form (pos='v' for verbs)
5. Remove stopwords - Filter out common words
6. `join()` - Combine back into text

This creates a clean, standardized corpus ready for vectorization.

```python
corpus = []

for i in range (0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    review = review.lower()
    review = word_tokenize(review)  # also use --> review.split()
    review = [lemma.lemmatize(word, pos='v') for word in review if word not in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)

corpus
```

## ✅ Corpus Created Successfully!
Now we have clean, preprocessed text ready for BOW transformation.

---

### Create Bag of Words

## 📊 CountVectorizer Parameters

| Parameter | Value | Purpose |
|-----------|-------|----------|
| `max_features` | 2500 | Keep only top 2500 most frequent words |
| `binary` | True | Use 1/0 (present/absent) instead of counts |

**Why binary?** For spam detection, word presence matters more than frequency.

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features = 2500, binary = True)

x = cv.fit_transform(corpus).toarray()
print(x.shape)
x
```

## 📖 Vocabulary Dictionary
Each word is mapped to a column index in the BOW matrix.

```python
cv.vocabulary_
```

---

## 🔗 N-grams: Capturing Word Pairs

**Problem with BOW:** "not good" and "very good" both contain "good" but mean different things.

**Solution: N-grams** - Combinations of consecutive words:
- **Unigram (1-gram)**: `["not", "good"]` 
- **Bigram (2-gram)**: `["not good"]`
- **Trigram (3-gram)**: `["not very good"]`

### Parameters:
- `ngram_range=(2, 3)` - Use both bigrams and trigrams
- `max_features=200` - Keep top 200 phrases
- `binary=True` - Track presence, not count

```python
## Create the Bag OF Words model with ngram
from sklearn.feature_extraction.text import CountVectorizer

cv1 = CountVectorizer(max_features=200, binary=True, ngram_range=(2, 3))

x1 = cv1.fit_transform(corpus).toarray()
print(x1.shape)
x1
```

## 📖 N-gram Vocabulary
Now our vocabulary contains word phrases like "free call", "urgent reply", etc.

```python
cv1.vocabulary_
```

## 🎯 Key Takeaways

✅ **BOW converts text to numbers** - Enables machine learning on text  
✅ **Preprocessing is critical** - Clean data → Better results  
✅ **Binary encoding** works well for presence-based tasks  
✅ **N-grams capture context** - Better than single words alone  
✅ **Trade-off**: More features = More accuracy but slower processing

---

### BOW Limitations to Remember:
- High dimensionality (thousands of features)
- No semantic understanding
- Loses word order completely
